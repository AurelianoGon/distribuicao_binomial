## Distribuição Binomial


## Nesta análise artificial na área da saúde foi aplicada a Distribuição Binomial no estudo da eficácia de uma nova vacina.

Um estudo foi simulado para testar a eficácia de uma nova vacina contra uma determinada doença. Simulou-se uma amostra de 100 pacientes e administrou-se a vacina a todos eles. Depois de um certo período, verificou-se quantos desses pacientes foram infectados pela doença.

Modelou-se o número de pacientes que não foram infectados (ou seja, que a vacina foi eficaz) como uma variável aleatória binomial.

* n (número de tentativas): O número de pacientes vacinados, que neste caso é 100.
* p (probabilidade de sucesso): A probabilidade de que um paciente vacinado não seja infectado. Se estudos anteriores sugerem que a vacina tem 95% de eficácia, então p = 0,95.
* X (variável aleatória): O número de pacientes na amostra que não foram infectados.

A Distribuição Binomial pode ser usada para calcular a probabilidade de um certo número de pacientes não serem infectados, ou para determinar o intervalo de confiança em que esperamos que a eficácia da vacina caia.

**Aplicação prática:**
Essa modelagem pode fazer inferências sobre a eficácia da vacina em uma população maior, estimando, por exemplo, a probabilidade de que pelo menos 90% dos pacientes em uma amostra maior também não sejam infectados, ou para comparar a eficácia desta nova vacina com outras disponíveis no mercado.

 Para calcular a probabilidade de um certo número de sucessos em uma distribuição binomial usei a **função binom.pmf (probability mass function) da biblioteca scipy.stats**

**A seguir, será calculado a probabilidade de exatamente 90 pacientes (de 100) serem infectados pela doença, assumindo que a eficácia da vacina é de 95%.**

```python
from scipy.stats import binom

# Parâmetros
n = 100   # número de pacientes vacinados
p = 0.95  # probabilidade de sucesso (eficácia da vacina)
k = 90    # número de sucessos desejados (pacientes não infectados)

# Calculando a probabilidade de exatamente 90 pacientes não serem infectados
prob = binom.pmf(k, n, p)

print(f"A probabilidade de exatamente {k} pacientes serem infectados é: {prob:.4f}")
```

**Calculando a probabilidade de pelo menos 90 pacientes não serem infectados, para isso foi somado as probabilidades para todos os valores de 90 até 100:**

```python
# Calculando a probabilidade de pelo menos 90 pacientes não serem infectados
prob_at_least_90 = binom.sf(89, n, p)  # sf é a função de sobrevivência (1 - cdf)

print(f"A probabilidade de pelo menos 90 pacientes não serem infectados é: {prob_at_least_90:.4f}")
```

 Em seguida, será visualizada as probabilidades de diferentes números de sucessos (pacientes não infectados) em uma amostra de 100 pacientes.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parâmetros
n = 100   # número de pacientes vacinados
p = 0.98  # probabilidade de sucesso (eficácia da vacina)

# Gerar uma gama de possíveis valores de k (número de pacientes não infectados)
k_values = np.arange(0, n+1)
probabilidades = binom.pmf(k_values, n, p)

# Plotando a distribuição binomial
plt.figure(figsize=(10, 6))
plt.bar(k_values, probabilities, color='darKgray')
plt.xlabel('Número de pacientes não infectados')
plt.ylabel('Probabilidade')
plt.title('Distribuição Binomial: Probabilidade de Pacientes Não Infectados\nEficácia da Vacina = 95%, n = 100')
plt.grid(True)
plt.show()
```
